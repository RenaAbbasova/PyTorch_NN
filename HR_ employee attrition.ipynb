### Handling outlier using IQR aproach

```python
df_num=df1.select_dtypes(include=['int','float'])
```

```python
df_num.head()
```

```python
df1.head()
```

## Capping approach

```python
Q3=df_num.quantile(0.75)
Q1=df_num.quantile(0.25)
IQR=Q3-Q1
Upper_fence=Q3+1.5*IQR
Lower_fence=Q1-1.5*IQR
print('upper_fence:\n', Upper_fence)
print('lower_fence:\n', Lower_fence)
```

```python
num_cols=['Age','DailyRate', 'DistanceFromHome','Education','EmployeeCount','EmployeeNumber','EnvironmentSatisfaction','HourlyRate','JobInvolvement','JobLevel','JobSatisfaction','MonthlyIncome','MonthlyRate','NumCompaniesWorked','PercentSalaryHike','PerformanceRating','RelationshipSatisfaction','StandardHours','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','WorkLifeBalance','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']
```

```python
df1[num_cols]=np.where(df1[num_cols]>Upper_fence,Upper_fence, np.where(df1[num_cols]<Lower_fence,Lower_fence,df1[num_cols]))
                        
```

```python
X = df1.drop('Attrition',axis=1)
y=df1['Attrition']
```

# Sequential Neural Network

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# Hyperparameters for our network
input_size = X.shape[1]
hidden_size = [128,64]
output_size = 2
```

```python
import torch
from torch import nn
# build i feed forward network
model = nn.Sequential(nn.Linear(input_size,hidden_size[0]),
                   nn.ReLU(),
                   nn.Linear(hidden_size[0],hidden_size[1]),
                   nn.ReLU(),
                   nn.Linear(hidden_size[1],output_size),
                   nn.Softmax(dim=1))
                   
print(model)                   
```

```python
# Define the loss
criterion=nn.NLLLoss()
```

```python
from torch import optim
```

```python
optimizer = optim.SGD(model.parameters(),lr=0.01)
```

```python
import torch.utils.data as Data
from torch import Tensor

X_train = Tensor(X_train)
y_train=Tensor(np.array(y_train))
```

```python
Batch_size=64

torch_dataset = Data.TensorDataset(X_train,y_train)#This line creates a PyTorch dataset using the TensorDataset class.

loader = Data.DataLoader(
        dataset = torch_dataset,
        batch_size =Batch_size,
        shuffle =True, num_workers=2,)#A data loader is responsible for creating batches of data from a dataset
```

# Training the model

```python
from torch.autograd import Variable

epochs=100
for i in range(epochs):
    running_loss = 0
    for step, (batch_x, batch_y) in enumerate(loader):
        
        # input node
        node_x = Variable(batch_x)# Variable is a node in computational graph. This stores data and graadient
        # output node
        node_y = Variable(batch_y.type(torch.LongTensor))
        
        # training pass
        optimizer.zero_grad()
        
        output = model(node_x)#
        loss = criterion(output,node_y)# how far actual 
        loss.backward()
        optimizer.step()# updating weight
        
        running_loss += loss.item()
        
    else:
        print(f'Training loss: {running_loss/len(X_train)}')
```

```python
X_test = Tensor(X_test)
y_test = Tensor(np.array(y_test))
z = model(X_test)
```
